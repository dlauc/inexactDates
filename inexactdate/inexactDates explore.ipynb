```python
from inexactdates import InexactDate
from datetime import datetime, timedelta
dayd, dayf = InexactDate('date', 1066, 10, 1).fset
print(dayd, len(dayf))
```

```python
y = [round(a*100,2) for a in dayf]
xp = datetime.fromordinal(dayd)
x = [(xp+timedelta(i)).strftime("%d.%m.%Y.") for i in range(len(dayf))]
[f'{a},{b}%' for a,b in list(zip(x,y))]
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

def inexactdateplot(date, l):
    dayd, dayf = date.fset
    y = [round(a*100,2) for a in dayf]
    xp = datetime.fromordinal(dayd)
    #x = [(xp+timedelta(i)).strftime("%d.%m.%Y.") for i in range(len(dayf))]
    x = [(xp+timedelta(i)) for i in range(len(dayf))]

    try:
        plt.close('all')
    except:
        pass
    fig, dax = plt.subplots(1)
    dax.plot(x, y, label=l)
    dax.legend(loc='upper left')

    fig.autofmt_xdate()

    dax.fmt_xdata = mdates.DateFormatter('%d.%m.%Y.')
    #plt.xticks(rotation=45)
    dax.set_xlabel("Datumi")
    dax.set_ylabel('Vjerojatnost članstva')
    #plt.legend(loc='upper left')
    fig.set_size_inches(18.5, 10.5)


    return plt
    

inexactdateplot(InexactDate('date', 1066, 10, 1), 'Približno 1. listopada 1066. godine').show()
```

```python
inexactdateplot(InexactDate('date', 1066, 10, 1, certainty=0.3), "Negdje oko 10. listopada 1066.").show()
```
